```python
#Employee Performance Bonus Eligibility
#create a empty dictionary name as employees
employees={}
n=int(input("Enter the number of Employees: "))
#Take employee name and score dynamically
for i in range(n):
  name=input("Enter the Employee Name: ")
  score=int(input("Enter the performance score: "))
  employees[name]=score
#to find highest score
highest_score=max(employees.values())
#find all employees with highest score
print("Top Performers Eligible for Bonus:",end=" ")
for name in employees:
  if employees[name]==highest_score:
    print(name,end=" ")
print("(Score:",highest_score,")")
```

```python
#Search Query Analysis
#Take search query input
query=input("Enter Search query: ")
#convert input into  lowercase
query=query.lower()
#Remove the  Punctuation
import string
for ch in string.punctuation:
  query=query.replace(ch,"")
#split into words
words=query.split()
#count the frequency of each word
count={}
for word in words:
  if word in count:
    count[word] +=1
  else:
    count[word]=1
#Display words searched more than once
result={}
for word in count:
  if count[word]>1:
    result[word]=count[word]
print(result)

```

```python
#Sensor Reading
#Take number of readings
n=int(input("Enter number of sensor readings: "))
sensor_reading=[]
#take readings
for i in range(n):
  value=int(input(f'Enter reading for hour {i}:'))
  sensor_reading.append(value)
#Store valid (enen) readings as (hour,value)
valid_readings=[]
for i in range(len(sensor_reading)):
  if sensor_reading[i]%2==0:
    valid_readings.append((i,sensor_reading[i]))
#Display Result
print("Valid Sensor Readings (Hour,Value):")
print(valid_readings)
```

```python
#Email Domain Usage Analysis
#take number of emails
n=int(input("Enter number of mails IDs: "))
emails=[]
#Take email inputs
for i in range(n):
  email=input("Enter email ID: ")
  emails.append(email)
#Count domain usage
domain_count={}
for email in emails:
  domain=email.split("@")[1] #Extract domain
  domain_count[domain]=domain_count.get(domain,0) +1
#calculate and display percentage
total =len(emails)
for domain in domain_count:
  percentage=(domain_count[domain]/total)*100
  print(domain + ":",str(int(percentage)) + "%")
```

```python
#Sales Spike Detection
#take number of days
n=int(input("Enter number of days: "))
sales=[]
#take daily sales input
for i in range(n):
  value=int(input(f"enter sales for days {i+1}: "))
  sales.append(value)
#calculate average
average=sum(sales)/len(sales)
#detect sales more than 30% above average
limit=average*1.30
print("sales spike days: ")
for i in range(len(sales)):
  if sales[i]>limit:
    print("Day",i+1,":",sales[i])
```

```python
#Duplicate User ID Detection
#take number of user IDs
n=int(input("Enter the number of user IDs: "))
user_ids=[]
#take input
for i in range(n):
  user_id=input(f"Enter user ID {i+1}: ")
  user_ids.append(user_id)
#count the occurences
count={}
for uid in user_ids:
  count[uid]=count.get(uid,0) + 1
#Display duplicates
print("Duplicates use IDs")
for uid in count:
  if count[uid]>1:
    print(f"{uid} -> {count[uid]} times")

```
