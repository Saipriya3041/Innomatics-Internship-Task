```python
#pass/fail analyzer
n=int(input("Enter the number of students"))
pass_count=0
fail_count=0
for i in range(n):
  marks=int(input(f'Enter the marks of {i+1}: student'))
  if marks>=50:
    pass_count +=1
  else:
    fail_count+=1
print("Total Number of students are:",n)
print("Total Number of Pass students are",pass_count)
print("Total Number of fail students are",fail_count)
```

```python
#Simple Data Cleaner
n=int(input("Number of names: "))
names=[]
for i in range(n):
  name=input(f'enter names {i+1}')
  names.append(name)
cleaned_names=[]
#remvoe the extra space in the list and covert to lowercase
for name in names:
  cleaned_name=name.strip().lower()
  cleaned_names.append(cleaned_name)
print("cleaned data")
for name in cleaned_names:
  print(name)


```

```python
#Message Length Analyzer
n=int(input("Enter the number of messages:"))
messages=[]
for i in range(n):
  message=input(f'Enter the messsages')
  messages.append(message)
for message in messages:
  length=len(message)
  print(f"Message: {message}")
  print(f'Length:{length}')
  if length>10:
    print("Flag: Message is longer than 10 characters")
print()
```

```python
#Error Messsage Detector
n=int(input("Enter the number of log entries:"))
logs=[]
for i in range(n):
  log=input(f'enter the log{i+1}')
  logs.append(log)
error_count=0
for log in logs:
  if "ERROR" in log.upper():
    error_count+=1
print(f"total number of log are {error_count}")
```
