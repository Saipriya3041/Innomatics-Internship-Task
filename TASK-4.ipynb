```python
#Mobile Recharge Validation System
def Validate_recharge(amount,valid_plans):
  if amount < 50:
    print("Recharge amount must be at least ₹50. ")
    return False
  elif amount not in valid_plans:
    print("Invalid plan selected")
    print("Available plans:",valid_plans)
    return False
  else:
    print("Recharge Successful!")
    return True
#take valid plans from user
valid_plans=list(map(int,input("Enter valid recharge plans (space separated):").split()))
#retry system using while loop
while True:
  try:
    amount=int(input("Enter recharge amount:  ₹"))
    if Validate_recharge(amount,valid_plans):
      break
    else:
      print("Please try again\n")
  except ValueError:
    print("Please enter a valid numeric amount.")

```

```python
#Inventory Reorder Alert System
def check_inventory(stock_data):
  for product,quantity in stock_data.items():
    if quantity <15:
      print(f"{product}:Recorder Alert(Stock={quantity})")
    else:
      print(f"{product}:Stock ok (Stock={quantity})")
#taking input
inventory={}
n=int(input("Enter number of products: "))
for i in range(n):
  product_name=input("Enter product name: ")
  quantity=int(input("Enter Stock quantity: "))
  inventory[product_name]=quantity
print("\n Inventory Status: \n")
check_inventory(inventory)
```

```python
#Student Result Processing System
def process_result(marks):
  total=0
  #calculate total using loop
  for mark in marks:
    total += mark
  average=total/len(marks)
  if average >=50:
    return average,"Pass"
  else:
    return average,"Fail"
#Taking input
marks_list=[]
n=int(input("Enter number of Subjects: "))
for i in range(n):
  mark=float(input(f"Enter marks for subject {i+1}: "))
  marks_list.append(mark)
#function call
average,result=process_result(marks_list)
print("\n----Result----")
print(f"Average Marks: {average:.2f}")
print(f"Status: {result}")
```

```python
#Cab Fare Estimator with Retry Option
def calculate_fare(distance,peak):
  base_fare=50
  per_km_rate=12
  fare=base_fare + (distance * per_km_rate)
  if peak.lower()=="yes":
    fare += fare *0.25 #25% extra
  return fare
#retry system using while loop
while True:
  try:
    distance=float(input("Enter distance traveled (in KM)"))
    peak=input("Is it peak hour? (yes/no)")
    if distance<=0:
      print("Distance must be greater than 0.\n")
      continue
    total_fare=calculate_fare(distance,peak)
    print(f"\n Total Fare: {total_fare:.2f}")
    retry=input("\nDo you want to calculate again? (yes/no)")
    if retry.lower() !='yes':
      print("Thank you for using Cab Fare Estimator!")
      break
    print()
  except ValueError:
    print("Please enter valid numeric distance.\n")
```

```python
#Employee Attendance Eligibility Checker
def check_eligibility(attendance_list):
  present_count=0
  #count present days using loop
  for status in attendance_list:
    if status.upper()=="P":
      present_count += 1
  total_days=len(attendance_list)
  if total_days ==0:
    return 0,"No Attendance Data"
  percentage=(present_count/total_days)*100
  if percentage >=75:
    return percentage, "Eligible"
  else:
    return percentage, "Not Eligible"
#take input
attendance=[]
n=int(input("Enter total working days: "))
for i in range(n):
  status=input(f"Enter attendace for day {i+1} (P/A): ")
  attendance.append(status)
#function call
percentage,result=check_eligibility(attendance)
print("\n-----Attendance Report-----")
print(f"Attendance Percentage: {percentage:.2f}%")
print(f"Status: {result}")
```

```python
#Password Strength Checker
def check_password_strength(password):
  has_digit=False
  has_special=False
  special_chars="@#$"
  #check each character using loop
  for char in password:
    if char.isdigit():
      has_digit=True
    if char in special_chars:
      has_special=True
  #Apply Conditions
  if len(password)<8:
    return "Weak Password (Minimum 8 characters required)"
  elif not has_digit:
    return "Weak Password (At least one 4 digit required)"
  elif not has_special:
    return "Weak Password (At least one special characers @#$ required)"
  else:
    return "Strong Password"
#taking input
while True:
  pwd=input("Enter your password: ")
  result=check_password_strength(pwd)
  print(result)
  retry=input("Do you want to try again? (yes/no): ")
  if retry.lower() != 'yes':
    break

```
