```python
# 1 Smart Light Controller
class SmartLight:
  def __init__(self,name):
    self.name=name
    self.status='OFF'#default status
  def turn_on(self):
    self.status='ON'
  def trun_off(self):
    self.status='OFF'
  def display_status(self):
    print(f'{self.name} is {self.status}')
#take input
light_name=input("Enter Light Name: ")
action=input("Enter action (ON/OFF): ").strip().upper()
#create object
light=SmartLight(light_name)
#performance action
if action =='ON':
  light.turn_on()
elif action=='OFF':
  light.trun_off()
else:
  print("Invalid action! please enter ON or OFF")
#display the result
print(light.display_status())
```

```python
# 2 Employee ID Card System
class Employee:
  def __init__(self,name,emp_id,department):
    self.name=name
    self.emp_id=emp_id
    self.department=department
  def display_id_card(self):
    print("\nEmployee ID Care")
    print(f"Name: {self.name}")
    print(f"ID: {self.emp_id}")
    print(f"Department: {self.department}")
#take input
name=input("Employee Name: ")
emp_id=input("Employee ID: ")
department=input("Department: ")
#create object
employee=Employee(name,emp_id,department)
#display ID card
employee.display_id_card()
```

```python
# 3 Mobile Contact Record
class Contact:
  def __init__(self,name,phone):
    self.name=name
    self.phone=phone
  def display_contact(self):
    print("\nContact Saved")
    print(f"Name: {self.name}")
    print(f"Phone Number: {self.phone}")
#take input
name=input("Enter Contact Name: ")
phone=int(input("Enter Phone Number"))
#create object
contact=Contact(name,phone)
#display contact information
contact.display_contact()
```

```python
# 4  Product Price Tag Generator
class Product:
  def __init__(self,name,price):
    self.name=name
    self.price=price
  def display_price_tag(self):
    print(f"\nProduct Name:{self.name}")
    print(f"Price: {self.price}")
#take input
product_name=input("Enter Product Nmae:  ")
price=int(input("Enter Prodcut Price ₹: "))
#create object
product=Product(product_name,price)
#display the Price Tag
product.display_price_tag()
```

```python
# 5 Movie Rating Display System
class Movie:
  def __init__(self,name,rating):
    self.name=name
    self.rating=rating
  def display_movie_details(self):
    print("\n Movie: ",self.name)
    print("Rating:",self.rating)
#take input
movie_name=input("Enter Movie Nmae: ")
rating=float(input("Rating"))
#create object
movie=Movie(movie_name,rating)
#display movie details
movie.display_movie_details()
```

```python
# 6 Delivery Address Manager
class Delivery:
  def __init__(self,customer_name,address):
    self.customer_name=customer_name
    self.address=address
  def display_delivery_details(self):
    print("\nDelivery Details")
    print(f"Customer Name: {self.customer_name}")
    print(f"Address: {self.address}")
#take input
customer_name=input("Enter Customer Name: ")
address=input("Enter Address: ")
#create object
delivery=Delivery(customer_name,address)
#display delivery details
delivery.display_delivery_details()
```
