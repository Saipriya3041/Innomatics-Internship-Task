```python
# 1 Smart Parking Lot Management system
def smart_parking_system():
  #taking input
  capacity=int(input("Enter parking capacity: "))
  logs_input=input("Enter Vehicle Logs separated by space (IN/OUT): ")
  #convert input string to list
  logs=logs_input.strip().split()
  current_parked=0
  peak_usage=0
  #process logs
  for log in logs:
    if log.upper()=='IN':
      current_parked +=1
    elif log.upper()=='OUT':
      if current_parked >0:
        current_parked-=1
    #Track peak usage
    if current_parked > peak_usage:
      peak_usage =current_parked
  #Capacity check
  if current_parked > capacity:
    print("Alert: Parking Exceeds Capacity! ")
  elif current_parked == capacity:
    print("Parking Status : Full")
  else:
    print("Parking Status: Available")
print(smart_parking_system())
```

```python
#2 Online Food Delivery Time Estimator
def estimate_delivery_time():
  #dynamic inputs
  distance=float(input("Distance(km)"))
  traffic=input("Traffic level (Low/Medium/High): ").strip().lower()
  weather=input("Weather (Clean/Rainy/Storm): ").strip().lower()
  #Base time calculation (let's say  30 km/h as average speed )
  base_time=(distance/30)*60 #convert hours to minutes
  #Traffic delay
  if traffic == "low":
    traffic_delay=5
  elif traffic=="medium":
    traffic_delay=15
  elif traffic=="high":
    traffic_delay=25
  else:
    traffic_delay=0
  #Weather delay
  if weather=="clear":
    weather_delay=0
  elif weather =="rainy":
    weather_delay=14
  elif weather =="storm":
    weather_delay=20
  else:
    weather_delay=0
  #Final ETA
  total_time=base_time+traffic_delay+weather_delay
  print("\n Estimated Delivery Time:",round(total_time),"minutes")
#call function
print(estimate_delivery_time())
```

```python
#3 Movie Theatre Seat Occupancy Analyzer
def seat_occupancy_analyzer():
  #dynamic input
  total_seats=int(input("Total Seats: "))
  booked_input=input("Enter booked seats as 1s separated by space: ")
  #convert input string into list
  booked_seats=list(map(int,booked_input.strip().split()))
  #calculate booked count
  booked_count=len(booked_seats)
  #calculate occupancy percentage
  occupancy=(booked_count/total_seats)*100
  print("\n-------Show Analysis------")
  print("Occupancy: ",round(occupancy),"%")
  #Determind show status
  if occupancy==100:
    print("Show Status: Housefull")
  elif occupancy>=75:
    print("Show Status: Almost Full")
  elif occupancy >=40:
    print("Show Status: Low Occupancy")
  #Suggest additional shows
  if occupancy>=90:
    print("Suggestion: Consider opening additional shows! ")
#call function
print(seat_occupancy_analyzer())
```

```python
#4 Cloud Server Load Classification System
def classify_server_load():
  #dynamic input
  readings_input=input("Enter CPU Readings (%) separated by space: ")
  #convert input string into list of interger
  readings=list(map(float,readings_input.strip().split()))

  #calculate total using loop
  total=0
  for value in readings:
    total += value
  #calculate average
  average=total/len(readings)
  print("\nAverage CPU Load:",round(average),"%")
  #classification
  if average<50:
    print("Server Status: Normal")
  elif 50<=average<80:
    print("Server Status: Warning")
  else:
    print("Server Status: Critical")
#call function
print(classify_server_load())

```

```python
#5 Smart Classroom Resource Usage Monitor
def classroom_resource_monitor():
  resources={}
  n=int(input("Enter number of resources: "))
  #Taking dynamic input for resource
  for i in range(n):
    name=input("Enter Resource Name: ")
    hours=float(input(f"Enter usage hours for {name}: "))
    resources[name]=hours
  overused=[]
  #check for overuse (threshold =8hours)
  for resource,usage in resources.items():
    if usage >8:
      overused.append(resource)
  print("\n----Resource Usage Report-----")
  if overused:
    print("Overused Resources:",",".join(overused))
    print("Enter Alert:yes")
  else:
    print("Overused Resource: None")
    print("Energy Alert: No")
#call the function
print(classroom_resource_monitor())
```

```python
#6  Online Event Registration Capacity Controller
def event_registration_controller():
  #taking input
  capacity=int(input("Event Capacity: "))
  total_registrations=int(input("Total Registrations Recived: "))
  confirmed=0
  waitlisted=0
  #Process registration using loop
  for i in range(total_registrations):
    if confirmed < capacity:
      confirmed+=1
    else:
      waitlisted +=1
  print("\n------Registration Summary------")
  print("Confirmed Registrations: ",confirmed)
  print("Waitlisted Users: ",waitlisted)
  #Registration status
  if confirmed >= capacity:
    print("Registration Status: Closed")
  else:
    print("Registration Status: Open")
#call the function
print(event_registration_controller())
```
