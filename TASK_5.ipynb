```python
#1.Social Media - Post Engagement Analyzer
def analyze_post_engagement():
  total_likes=0
  #Ask user how many entries they want to add
  n=int(input("Enter number of like entries: "))
  for i in range(n):
    likes=int(input(f"Enter likes for entry {i+1} "))
    total_likes += likes
  #Determine post status
  if total_likes >= 1000:
    status="Viral Post"
  else:
    status="Normal Engagement"
  #Display results
  print(f"Total Likes:{total_likes}")
  print(f"Post Status: {status}")
#Call the function
analyze_post_engagement()
```

```python
#3.Agriculture - Rainfall Adequacy Checker
def rainfall_checker():
  total_rainfall=0
  #number of months/seasons
  n=int(input("Enter number of rainfall records: "))
  #loop to cellect rainfall data
  for i in range(n):
    rainfall=float(input(f"Enter rainfall for period {i+1} "))
    total_rainfall += rainfall
  #calculate average rainfall
  average_rainfall=total_rainfall/n
  #required rainfall leve
  required_level=float(input("Enter required average rainfall level: "))
  #check adequacy
  if average_rainfall >= required_level:
    status="Adequate Rainfall"
  else:
    status="Inadequate Rainfall"
  #display results
  print(f"\n Average Rainfall: {round(average_rainfall,2)} ")
  print(f"Rainfall status:{status} ")
#call the function
rainfall_checker()
```

```python
#4 Social Media-Duplicate Account Detection
def detect_duplicate_accounts():
  usernames=[]
  #take input
  n=int(input("Enter usernames: "))
  for i in range(n):
    username=input(f"Enter Username {i+1}")
    usernames.append(username)
    #use set to detect duplicate
    unique_usernames=set(usernames)
  if len(unique_usernames) <len(usernames):
    result="Yes"
    print(f"\nDuplicate Accounts Found: {result}")
  else:
    result="No"
    print(f"\n Duplicate Accounts Found: {result}")
#call the function
detect_duplicate_accounts()
```

```python
#6.Agriculture - Premium Crop Price Filter
def filter_premium_cross(prices):
  premium=[price for price in prices if price >2000]
  return premium
#taking input
crop_prices=list(map(int,input("Enter crop prices separated by space: ").split()))
premium_crops=filter_premium_cross(crop_prices)
print("Premium Crops: ",premium_crops)
```

```python
#7.System Monitoring - Application Health Checker
def check_system_health(error_count):
  if error_count==0:
    return "Healthy"
  elif error_count<=5:
    return "Minor Issues"
  else:
    return "Critical Issues"
#take input
errors=int(input("Enter Error Count: "))
status=check_system_health(errors)
print("Error Count: ",errors)
print("System Status: ",status)
```

```python
#8.Banking - Daily Transaction Limit Checker
def check_transaction(amount):
  daily_limit=5000
  if amount<=daily_limit:
    return "Approved"
  else:
    return "Rejected"
#take input
transaction_amount=int(input("Enter Transaction Amount: "))
status=check_transaction(transaction_amount)
print("Transaction Amount: ",transaction_amount)
print("Transaction Status: ",status)
```

```python
#9.Learning - Student Attendance Eligibility System
def check_attendance(attendance_list):
  total_classes=len(attendance_list)
  attended=0
  for status in attendance_list:
    if status == 1:
      attended+=1
  percentage=(attended/total_classes)*100
  if percentage >=75:
    eligibility ="Eligible"
  else:
    eligibility="Not Eligible"
  return percentage,eligibility
#take input
print("Enter Attendance record(1 for Present , 0 for Absent)")
attendance=list(map(int,input("Separate by space: ").split()))
percentage,status=check_attendance(attendance)
print("Attendance Percentage: ",round(percentage,1))
print("Exam Eligibility: ",status)
```

```python
#10.Smart Electricity Bill Analyzer
def calculate_bill(units):
  bill=0
  #Billing
  if units<=100:
    bill=units*3
  elif units<=200:
    bill=(100*3)+((units-100)*5)
  else:
    bill=(100*3)+(100*5)+((units-200)*7)
  #Usage Classification
  if bill<=500:
    usage="Low Usage"
  elif bill<=1500:
    usage="Moderate Usage"
  else:
    usage="High Usage"
  return bill,usage
#take input
units_consumed=int(input("Enter units consumed: "))
total_bill,usage_status=calculate_bill(units_consumed)
print("Total Bill:₹ ",total_bill)
print("Usage Level: ",usage_status)
```
